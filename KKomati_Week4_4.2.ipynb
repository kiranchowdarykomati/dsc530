```python
import numpy as np
import nsfg
import first
import thinkstats2
import thinkplot

# Generate 1000 random numbers
r = np.random.random(1000)

# Calculate pmf
pmf = thinkstats2.Pmf(r)

# Plot pmf
thinkplot.Pmf(pmf, linewidth=0.1)
thinkplot.Config(xlabel='Generated Random Value', ylabel='PMF')
```

```python
# Calculate cdf
cdf = thinkstats2.Cdf(r)

# Plot cdf
thinkplot.Cdf(cdf)
thinkplot.Config(xlabel='Generated Random Value', ylabel='CDF')
```
